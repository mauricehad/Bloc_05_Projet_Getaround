```python
# Basic stats
print("Number of rows : {}".format(delay.shape[0]))
print()

print("Display of delay: ")
display(delay.head())
print()

print("Basics statistics: ")
data_desc = delay.describe(include="all")
display(data_desc)
print()

print("Percentage of missing values: ")
display(100 * delay.isnull().sum() / delay.shape[0])
```
